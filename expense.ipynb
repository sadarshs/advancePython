```python
def add_expense():
    try:
        c = input("Category: ")
        a = float(input("Amount: "))
        open("expenses.txt", "a").write(f"{c},{a}\n")
    except:
        print("Invalid amount!")


def view_expenses():
    try:
        for line in open("expenses.txt"):
            c, a = line.strip().split(",")
            print(c, a)
    except:
        print("File not found!")


def calculate_total():
    total = 0
    try:
        for line in open("expenses.txt"):
            _, a = line.strip().split(",")
            total += float(a)
        print("Total:", total)
    except:
        print("File not found!")


while True:
    print("\n1.Add 2.View 3.Total 4.Exit")
    ch = input("Choice: ")

    if ch == "1":
        add_expense()
    elif ch == "2":
        view_expenses()
    elif ch == "3":
        calculate_total()
    elif ch == "4":
        break
    else:
        print("Invalid!")
```
